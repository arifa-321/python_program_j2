# Iterations in python 
1.Iterations are used to work on repeatative tasks or repeatative performances 
2.python having 2 types of interations 
3. `for` loop
4. `while` loop 

```python - for loop 
for iter_variables in sequence:
    print("for is keyword to start the iterarion")
    print("iter_variables is to get values from sequence") 
    print("in is keyword to valid sequence present or not")
    print("sequence is a collection like range,string,list,tuple...etc")
```

```python - while loop 
while (condition): 
    print("while is keyword to valid condition based on data or values")
    print("if condition is True then return inside statements")
```

````python
# write a program to print 1 to 20 numbers using for loop 
for i in range(1,21): 
    print(i)
````

````python
# write a program to iterae string using for loop 
info="welcome to python program"
for char in info:
    print(char)
````

````python
# write a program to print even numbers in between 50 to 100 using for loop?
for n in range(52,101):
    if (n%2 == 0):
        print(n,end=",")
````

````python
# write a program to print odd numbers in between 50 to 100 using for loop?
for n in range(50,101):
   if (n%2 == 1):
     print(n,end=",")
````

````python
# write a program to find the vowels from string using for loop?
data="python programming"
for char in data:
    if(char in "aeiouAEIOU"):
      print(char,end=",")
````

````python
# write a program to find the vowels from string using for loop?
data="python programmimg"
for char in data:
    if(char not in "aeiouAEIOU"):
       print(char,end=",")
````
