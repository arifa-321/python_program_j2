3. `elif` statement
4. `else` statement
5. `nested` statement

```python syntax - if

    statement1 
    statement2...
```

```python - if, else 
if (condition):
    statement1
    statement2...
else:
    statement1
    statement2..
```

````python
""" 
write a program to find the grades based on marks 
1. if marks in between 80-100 - grade is A 
2. if marks in between 60-79 - grade is B
3. if marks in between 40-59 - grade is C 
4. if marks in between 25-39 - grade is D 
5. if marks in between 0-25 - grade is Fail 
6. if marks other than above output is - please enter correct marks 
"""
marks=int(input("please enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks,"grade is A")
elif (marks>=60 and marks<=79):
   print(marks,"grade is B")
elif (marks>=40 and marks<=59):
   print(marks,"grade is C")
elif (marks>=25 and marks<=0):       
   print(marks,"grade is D")
elif (marks>=0 and marks<=24):   
   print(marks,"grade is Fail")
else:
   print("please enter correct marks")
````
