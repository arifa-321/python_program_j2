# operators in python 

## 6 types of operations in python 
1. arithematic operations
2. assignment operators
3. comparision operators
4. logical operators
5. membership operators
6. identity operators 

```python
# arithematic operators -,+,*,/,//,% 
a=10
b=20
c=20
d=30
print(a+b)
print(b-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)
```

```python
#assingment operators - +=,-=,*=,/=,//=,%=
a=100
a+=50 
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c=50
c*=100
print(c)
```

```python
#comparition operators - <,>,=,<=,>=,!=
a=10
b=20
c=20
d=30
print(a>b)
print(b==c)
print(b<c)
print(b!=c)
print(b>=a)
print(b<=a)
```

```python
#logical operators - and,or,not 
"""
and - both conditons are true then result is true , rest case is false 
or - either one condition is true then the result is true, rest case is false
and - truth table 
T T = T
T F = F
F T = F
F F = F 
or - truth table 
T T = T 
T F = T 
F T = T 
F F = F 
"""
a=10
b=20
c=20
d=30
print(a>b and b==c)
print(a<b and c==b) 
print(a>b or b==c)  
print(not a>b and b==c)
```

```python
#membership operations - in,not in 
# volidating members are presnt in that sequence or group 
names=["vikas","shivaji","sidh","chandra"]
print("vikas" in names)
print("shivji" not in names)
print("Vikas" not in names)
print("vikram" in names) 
```

```python
# identity operations - is, is not
a=10000
b=10000
print(a is b)
print(id(a))
print(id(b))
```

```python
a=10
b=10 
print(a is b)
print(id(a))
print(id(b))
```
